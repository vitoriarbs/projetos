Case - Cancelamento de Clientes

Recentemente a empresa percebeu que da sua base total de clientes, grande maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Nessa análise vamos desenvolver uma pesquisa para entender os principais motivos desses cancelamentos e quais as açoes mais eficientes para reduzir esse número.

# Passo a Passo:
# Passo 1: Importar a base de dados utilizando pandas
# Passo 2: Visualizaçao da base de dados (entender a base + identificar problemas)
# Passo 3: Tratamento dos dados da tabela
# Passo 4: Análise
#   Quantos clientes cancelaram?
#   Qual o $ de clientes que cancelaram?
# Passo 5: Análise da causa de cancelamento
#   Comparar outras colunas com a coluna cancelou

```python
# Passo 1: Importar a base de dados utilizando pandas
import pandas as pd
tabela = pd.read_csv("cancelamentos.csv")

#Passo 2:Visualização da base de dados
tabela = tabela.drop(columns="CustomerID") #iremos retirar a variável CustomerID da visualização, pois não será útil para a análise
display(tabela)
```

```python
#Passo 3: Tratamento dos dados da tabela
display(tabela.info()) #para visualizar as informações da tabela, teremos 881666 valores iniciais

#deletando valores vazios
tabela = tabela.dropna() #NaN is not a number = valores vazios
display(tabela.info()) #teremos 881659 valores, nenhum vazio
```

```python
#Passo 4: Análise

#contagem de quantos clientes cancelaram
cancelou = tabela["cancelou"].value_counts()
display(cancelou)

# 0.0 = True - Não cancelaram
# 1.0 =  False - Cancelaram

# % de clientes que cancelaram
display(tabela["cancelou"].value_counts(normalize=True))
```

```python
#Passo 5: Análise da causa de cancelamento

#comparar as outras colunas da tabela com a coluna de concelamento

import plotly.express as px #plotly será utilizado para a visualizaçao de gráficos e comparação de colunas

#criaçao do gráfico
for coluna in tabela.columns:
    grafico = px.histogram(tabela, x="duracao_contrato", color="cancelou", text_auto=True)
    #exibição do gráfico
    grafico.show()
```
